```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib

import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import skew
from scipy.stats import norm
from scipy.stats.stats import pearsonr
```

```python
data_type = {
    'StoreID': int,
    'Date':str,
    'IsHoliday':bool,
    'IsOpen':bool,
    'HasPromotions':bool,
    'StoreType':'category',
    'AssortmentType':'category',
    'NearestCompetitor':int,
    'Region':int,
    'NumberOfCustomers':int,
    'NumberOfSales':int,
    'Region_AreaKM2':float,
    'Region_GDP':float,
    'Region_PopulationK':int,
    'CloudCover':float,
    'Events':'category',
    'Max_Dew_PointC':float,
    'Max_Gust_SpeedKm_h':float,
    'Max_Humidity':float,
    'Max_Sea_Level_PressurehPa':float,
    'Max_TemperatureC':float,
    'Max_VisibilityKm':float,
    'Max_Wind_SpeedKm_h':float,
    'Mean_Dew_PointC':float,
    'Mean_Humidity':float,
    'Mean_Sea_Level_PressurehPa':float,
    'Mean_TemperatureC':float,
    'Mean_VisibilityKm':float,
    'Mean_Wind_SpeedKm_h':float,
    'Min_Dew_PointC':float,
    'Min_Humidity':float,
    'Min_Sea_Level_PressurehPa':float,
    'Min_TemperatureC':float,
    'Min_VisibilitykM':float,
    'Precipitationmm':float,
    'WindDirDegrees':float
    } 

train = pd.read_csv("train.csv",dtype=data_type,parse_dates=["Date"],dayfirst=True)
test = pd.read_csv("test.csv",dtype=data_type,parse_dates=["Date"],dayfirst=True)

#train.set_index(["Date","StoreID"],inplace=True)
#test.set_index(["Date","StoreID"],inplace=True)
```

```python
all_data = pd.concat((train.drop('NumberOfSales',axis=1),test))
all_data.describe()
```

```python
matplotlib.rcParams['figure.figsize'] = (8.0, 6.0)
# fit the data with a normal distribution and 
sns.distplot(train['NumberOfSales'] , fit=norm)

# and check the fitted mu and sigma
(mu, sigma) = norm.fit(train['NumberOfSales'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))
print("Skewness: %f" % train['NumberOfSales'].skew())
print("Kurtosis: %f" % train['NumberOfSales'].kurt())

plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('NumberOfSales distribution')
```

```python
all_data_na = (all_data.isnull().sum() / len(all_data)) * 100
all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values(ascending=False)[:30]
missing_data = pd.DataFrame({'Missing Ratio' :all_data_na})
missing_data.head(10)
```

```python
# IMPUTATION
all_data.Events = all_data.Events.fillna('Normal')

all_data['Max_Gust_SpeedKm_h'] = all_data['Max_Gust_SpeedKm_h'].fillna(all_data['Max_Gust_SpeedKm_h'].mode()[0])
all_data['CloudCover'] = all_data['CloudCover'].fillna(all_data['CloudCover'].mode()[0])
all_data['Min_VisibilitykM'] = all_data['Min_VisibilitykM'].fillna(all_data['Min_VisibilitykM'].mode()[0])
all_data['Mean_VisibilityKm'] = all_data['Mean_VisibilityKm'].fillna(all_data['Mean_VisibilityKm'].mode()[0])
all_data['Max_VisibilityKm'] = all_data['Max_VisibilityKm'].fillna(all_data['Max_VisibilityKm'].mode()[0])
all_data = all_data.drop('NumberOfCustomers',axis=1)

train['Max_Gust_SpeedKm_h'] = train['Max_Gust_SpeedKm_h'].fillna(train['Max_Gust_SpeedKm_h'].mode()[0])
train['CloudCover'] = train['CloudCover'].fillna(train['CloudCover'].mode()[0])
train['Min_VisibilitykM'] = train['Min_VisibilitykM'].fillna(train['Min_VisibilitykM'].mode()[0])
train['Mean_VisibilityKm'] = train['Mean_VisibilityKm'].fillna(train['Mean_VisibilityKm'].mode()[0])
train['Max_VisibilityKm'] = train['Max_VisibilityKm'].fillna(train['Max_VisibilityKm'].mode()[0])

all_data_na = (all_data.isnull().sum() / len(all_data)) * 100
all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values(ascending=False)[:30]
missing_data = pd.DataFrame({'Missing Ratio' :all_data_na})
missing_data.head(10)
```

```python
categorical_variables_mask = train.dtypes==object
categorical_variables = train.columns[categorical_variables_mask].tolist()
int_vars_mask = train.dtypes == int
float_vars_mask = train.dtypes == float
num_vars_mask = int_vars_mask | float_vars_mask
num_vars = train.columns[num_vars_mask].tolist()
num_vars = num_vars[3:]
```

```python
norm_train = train.copy()
norm_train[num_vars] = (norm_train[num_vars] - norm_train[num_vars].mean())/norm_train[num_vars].std()
```

```python
cov=norm_train[num_vars].corr(method='pearson')
plt.figure(figsize = (16,16))
sns.heatmap(cov,square=True,annot=True,cmap="Blues",cbar=False,xticklabels=True, yticklabels=True)
```
